```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style='whitegrid', font_scale=1.5)
sns.set_palette('Set2', n_colors=10)
plt.rc('font', family='malgun gothic')
plt.rc('axes', unicode_minus=False)

import streamlit as st
from datetime import date
```

```python
job_2022 = pd.read_csv('국세청_사업자현황_100대 생활업종_20220228.csv',encoding='euc-kr')
job_2021 = pd.read_csv('국세청_사업자현황_100대 생활업종_20210131.csv',encoding='euc-kr')
gen = pd.read_csv('국세청_사업자현황_성별_20220228.csv', encoding='euc-kr')
age = pd.read_csv('국세청_사업자현황_연령별_20220228.csv', encoding='euc-kr')
yea = pd.read_csv('국세청_사업자현황_존속연수별_20220228.csv', encoding='euc-kr')
```

```python
# 서울지역으로만 한정해 데이터 뽑기

job_2021_seoul = job_2021[job_2021['시도'] == '서울특별시']
job_2022_seoul = job_2022[job_2022['시도'] == '서울특별시']

# *로 처리된 결측치 0으로 바꿔주기

job_2021_seoul = job_2021_seoul.replace('*', np.nan)
job_2021_seoul = job_2021_seoul.replace(' * ', np.nan)

job_2022_seoul = job_2022_seoul.replace('*', np.nan)

job_2021_seoul.fillna(0, inplace = True)
job_2022_seoul.fillna(0, inplace = True)
```

```python
# 실생활과 밀접한 직종만 뽑기

types = ['LPG 충전소',
 'PC방',
 '가구점',
 '가전제품수리점',
 '가전제품판매점',
 '간이주점',
 '간판광고물업',
 '건어물가게',
 '곡물가게',
 '과일가게',
 '교습소ㆍ공부방',
 '교습학원',
 '기타음식점',
 '꽃가게',
 '노래방',
 '당구장',
 '독서실',
 '목욕탕',
 '문구점',
 '미용실',
 '부동산중개업',
 '분식점',
 '사진촬영업',
 '생선가게',
 '서점',
 '세탁소',
 '슈퍼마켓',
 '스포츠시설운영업',
 '스포츠용품점',
 '시계ㆍ귀금속점',
 '식료품가게',
 '신발가게',
 '실내스크린골프점',
 '실내장식가게',
 '실외골프연습장',
 '안경점',
 '애완용품점',
 '여관ㆍ모텔',
 '여행사',
 '예식장',
 '옷가게',
 '이륜자동차판매점',
 '이발소',
 '일식전문점',
 '자동차수리점',
 '자전거판매점',
 '장난감가게',
 '정육점',
 '제과점',
 '주유소',
 '주차장운영업',
 '중고차판매점',
 '중식전문점',
 '채소가게',
 '철물점',
 '침구ㆍ커튼가게',
 '커피음료점',
 '컴퓨터판매점',
 '통신판매업',
 '패스트푸드점',
 '펜션ㆍ게스트하우스',
 '편의점',
 '피부관리업',
 '한식전문점',
 '헬스클럽',
 '호프전문점',
 '화장품가게',
 '휴대폰가게']

job_2021_fit = job_2021_seoul.loc[job_2021_seoul['업종'].isin(types)]
job_2022_fit = job_2022_seoul.loc[job_2022_seoul['업종'].isin(types)]
```

```python
# 데이터 타입 int로 변경

job_2021_fit['당월'] = job_2021_fit['당월'].astype(int)
job_2021_fit['전월'] = job_2021_fit['전월'].astype(int)
job_2021_fit['전년동월'] = job_2021_fit['전년동월'].astype(int)

job_2022_fit['당월'] = job_2022_fit['당월'].astype(int)
job_2022_fit['전월'] = job_2022_fit['전월'].astype(int)
job_2022_fit['전년동월'] = job_2022_fit['전년동월'].astype(int)
```

```python
# 열 이름 일자로 변경

col_1 = {'당월':'2021_1', '전년동월':'2020'}
job_2021_fit = job_2021_fit.rename(columns = col_1)

col_2 = {'당월':'2022', '전년동월':'2021_2'}
job_2022_fit = job_2022_fit.rename(columns = col_2)
```

```python
# 업종과 시도를 기준으로 groupby

job_2021_fit = job_2021_fit.groupby(['업종','시도'])[['2021_1','전월','2020']].sum()
job_2022_fit = job_2022_fit.groupby(['업종','시도'])[['2022','전월','2021_2']].sum()
```

```python
# 2020년부터 2022년까지 병합 

job_3 = pd.merge(job_2021_fit, job_2022_fit, on='업종', how='left')

coll = ['2020', '2021_1','2021_2', '2022']
job = job_3.loc[:,coll]
```

```python
#  2020~2021  /  2021~2022  /   2020~2022  각 년도별 증감추이 

job['2020-2021 증감%'] = ((job['2021_1'] - job['2020'])/job['2020'])*100
job['2021-2022 증감%'] = ((job['2022'] - job['2021_2'])/job['2021_2'])*100
job['2년간 증감%'] = ((job['2022'] - job['2020'])/job['2020'])*100
```
